```python
import caas
import caas.lib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
len(caas.color_definitions)
```

```python
rgb = caas.color_definitions['ral_standard']['rgbs']
len(rgb)
type(rgb)
np.shape(rgb)
```

```python
%matplotlib notebook
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(rgb[:,0],rgb[:,1],rgb[:,2], c=rgb/256, alpha=1)

ax.set_xlabel('red')
ax.set_ylabel('green')
ax.set_zlabel('blue')

plt.show
```
